- 111... is always largest which are $2^g$
- Numbers with the tail (01)+ are the same as the number without the tail.  They keep moving toward the center of the sort order as generations grow.
- Every "capital G' Generation will contain the _g_ powers of 2 <= $2^g$
    - 010101... is always 1: I think its tree indexes from the top of each generation are $[0, 1 (01), 6 (0110), ]$ starting with generation 0?? then g+2, g+4, ...
    - 1010101... is always 2: ??? Indexes from generation top will be same as 1 but starting from generation 2 ???
    - 11010101... is always 4
    - etc
- [1*]000 for generations >=3 is always the minimum value (three zeros in a row turns a large positive number into a large negative number)
    - The smallest values are: $-\ \large{\frac{11 \cdot 6^{(g-3)}}{3^{g}}}$ for generation, g, greater than 2
 
The goal is to be able to translate any Collatz Path to its corresponding value index in its generation.

Sort order is going to be driven by some parameters:
- Tail that makes it negative or positive
- Starting one count of path
- Ratio of ones to zeros in the path

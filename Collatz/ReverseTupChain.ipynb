# Tup with S-List is reversable
Now that we are carrying S-List in the Node tuple we can always reverse a tuple

```python
import sys, io
import math
import numpy as np
import pandas as pd
from scipy.optimize import nnls
from fractions import Fraction
from sympy import factorint
from itertools import product
```

```python
tup_0 = (0, 0, 0, [], 0)

def F_0(tup):
    a,b,c,S,p = tup
    return (a+1, b+1, (3*c) + (2**(a+p)), S+[a+p], p^1)
#
def F_1(tup):
    a,b,c,S,p = tup
    return (a+1, b, c, S, p^1)
#
def s_Sum(S):
    m = len(S)
    total = 0
    for r in range(m):
        total += ( (3**(m-1-r)) * (2**(S[r])) )
    return total
#

def F_r(tup):
    """
    Reverses an F_0 or F_1 operation
    """
    a,b,c,S,p = tup
    if (a == 0):
        return None
    len_S = len(S)
    if len_S > 0:
        max_zero = S[-1]
        if max_zero != (a-1):
            return (a-1, b, c, S, p^1)
        else:
            if len_S > 1:
                S = S[0:-1]
            else:
                S = []
            c = s_Sum(S)            
            return(a-1, b-1, c, S, p^1)
    else:
        return (a-1, b, c, S, p^1)
#
```

```python
def ChainPath(collatzNumber):
    path = []
    while collatzNumber != 1:
        if (collatzNumber & 1) == 0:
            collatzNumber = collatzNumber // 2
            path.append("1")
        else:
            collatzNumber = (3 * collatzNumber + 1) // 2
            path.append("0")
    return "".join(path)
#
def TupChainFromPath(chain_path):
    path_len = len(chain_path)
    tup_chain = [(0, 0, 0, [], 0)]
        
    for chain_item in chain_path:
        if chain_item == "1":
            tup_chain.append(F_1(tup_chain[-1]))
        else:
            tup_chain. append(F_0(tup_chain[-1]))
    #
    return tup_chain
#
def ReverseTupChainFromTup(tup):
    L = [tup]
    while(tup[0] > 0):
        tup = F_r(tup)
        L.append(tup)
    return L
#
```

```python
tup_chain = TupChainFromPath(ChainPath(7))
reverse_tup_chain = ReverseTupChainFromTup(tup_chain[-1])
tup_chain, reverse_tup_chain
```

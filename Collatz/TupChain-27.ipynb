I have a series of triplets of integers a, b, c such that:
 a_(i+1) = a_(i) + 2
 b_(i+1) = b_(i) + 1
 c_(i+1) = 3 * c(i) + 2^(a_i)
 and for all i:
 D = (2^(a_(i)) - c_(i)) / 3^(b_(i))

Given any integer D, I want to find integers a_(0), b_(0), c_(0).

For instance for 5 we have a_(0) = 4, b_(0) = 1, c_(0) = 1 and a_(1) = 6, b_(1) = 2, c_(1) = 19

Generate an algorithm for finding integer a_(0), b_(0), c_(0) given an integer D


$f(x)=\dfrac{3x+1}{2}$ is affine: $x\mapsto ax+b$ with $a=\tfrac32,\ b=\tfrac12$.

Composing $N$ times gives

$$
f^{\,N}(x)=a^{N}x+b\sum_{k=0}^{N-1}a^{k}
= a^{N}x+b\,\frac{a^{N}-1}{a-1}.
$$

Here $\dfrac{b}{a-1}=\dfrac{1/2}{1/2}=1$, so

$$
\boxed{\,f^{\,N}(x)=\left(\tfrac32\right)^{\!N}(x+1)-1
= \frac{3^{N}x+3^{N}-2^{N}}{2^{N}}\,}
$$


```python
def pathMeta(chain_path):
    R = {"0": 0, "1": 0, "NOOP": 0} 
    n = len(chain_path)
    i = 0
    while i < n:
        if i < n-1:
            if chain_path[i:i+2] == "01":
                R["NOOP"] += 1
                i += 2
                continue
        if chain_path[i] == "0":
            R["0"] += 1
        else:
            R["1"] += 1
        i += 1
    #
    x = 1
    n = R["0"]
    m = R["1"]
    lower = (((3/2)**n)*(x + 1) - 1)/(2**m) 
    # upper isn't right ...
    n = R["0"] + R["NOOP"]
    m = R["1"] 
    upper = (((3/2)**n)*(x + 1) - 1)/(2**m)
    
    return (R, lower, upper)
#
```
